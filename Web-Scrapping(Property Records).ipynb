```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
props = requests.get('https://www.propertypro.ng/property-for-rent')
```

```python
props
```

```python
props.content
```

```python
rent = BeautifulSoup(props.content, 'html.parser')
```

```python
print(rent.prettify())
```

```python
r1 = rent.find_all('h3', {'class':'listings-property-title2'}) #Title
```

```python
len(r1)
```

```python
r2 = rent.find_all('h3', {'class':'listings-price'}) #Price
```

```python
len(r2)
```

```python
r3 = rent.find_all('h4', {'class':'listings-property-title'}) #Description
```

```python
len(r3)
```

```python
r4 = rent.find_all('div', class_='phone-icon') #Phone Numbers
```

```python
len(r4)
```

```python
#TO SEE ALL THE PHONE NUMBERS
for x in r4:
    print(x.text)
```

```python
rr4 = r4[:50] #Picking the first 50 number
```

```python
len(rr4)
```

```python
for x in r1:
    print(x.text)
```

```python
for x in r2:
    print(x.text)
```

```python
for x in r3:
    print(x.text)
```

```python
#WE WANT TO DROP THE CONTENT OF r1 IN TITLE LIST
title = []
for x in r1:
    title.append(x.text)
```

```python
#WE WANT TO DROP THE CONTENT OF r2 IN PRICE LIST
price = []
for x in r2:
    price.append(x.text)
```

```python
#WE WANT TO DROP THE CONTENT OF r3 IN DESCRIPTION LIST
description = []
for x in r2:
    description.append(x.text)
```

```python
#WE WANT TO DROP THE CONTENT OF rr4 IN PRICE LIST
contact = []
for x in rr4:
    contact.append(x.text.strip())
```

```python
#WE CREATE A DICTIONARY NAMED property
property = {'Property Title':title, 'Property Description':description, 'Annual Rent':price, 'Contact Phone':contact}
```

```python
import pandas as pd
```

```python
#WE CONVERT THE PROPERTY DICTIONARY TO A DATAFRAME
property = pd.DataFrame(property)
```

```python
property
```

```python
property.to_csv('Property Records.csv', index=False)
```
